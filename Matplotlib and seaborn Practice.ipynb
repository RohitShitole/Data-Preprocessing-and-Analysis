```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.array([1,2,3,4,5,6])
y=np.array([10,20,40,50,60,70])
plt.plot(x,y,color='r')
#plt.subplot(2,3,1)
plt.show()
```

```python
x=np.array([2,3,4,5,6])
y=np.array([10,20,30,40,70])
font={'family':'fantasy','color':'r','size':'20'}
font1={'family':'serif','color':'g','size':'30'}
plt.title("Data Science",fontdict=font)
plt.xlabel("name",fontdict=font1)
plt.ylabel("age",fontdict=font)
plt.plot(x,y,marker='p')
plt.show()
```

```python
x=np.array([2,3,4,5,6])
y=np.array([10,20,30,40,70])
font={'family':'fantasy','color':'r','size':'20'}
font1={'family':'serif','color':'g','size':'30'}
plt.title("Data Science",fontdict=font)
plt.xlabel("name",fontdict=font1)
plt.ylabel("age",fontdict=font)
plt.bar(x,y)
plt.show()
```

```python
y=np.array([10,20,30,40,70])
plt.pie(y,labels=y)
plt.show()
```
